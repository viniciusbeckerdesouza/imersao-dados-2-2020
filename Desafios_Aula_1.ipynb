```
import pandas as pd

fonte = "https://github.com/alura-cursos/imersao-dados-2-2020/blob/master/MICRODADOS_ENEM_2019_SAMPLE_43278.csv?raw=true"

dados = pd.read_csv(fonte)
```

```
provas = ["NU_NOTA_CN","NU_NOTA_CH","NU_NOTA_MT","NU_NOTA_LC","NU_NOTA_REDACAO"]
```

#Desafio01: Proporção dos inscritos por idade.

```
idade = dados["NU_IDADE"].value_counts().sort_index()
total = dados["NU_IDADE"].count()

for i in idade:
    for d in dados['NU_IDADE']:
        proporcao=((i/total)*100)
        print("%d = %f " % (d, proporcao))
        



```

#Desafio03: Adicionar título no gráfico

```
import matplotlib.pyplot as plt
dados[provas].boxplot(grid=True, figsize= (10,8))
plt.title('Boxplots das Notas do ENEM')
```

#Desafio04: Plotar os Histogramas das idades dos treineiros e não treineiros.

```
treineiros = dados.query('IN_TREINEIRO==1')
plt.hist(treineiros['NU_IDADE'], bins = 20)
plt.title("Idades dos Treineiros")
```

```
nao_treineiros = dados.query('IN_TREINEIRO==0')
plt.hist(nao_treineiros['NU_IDADE'], bins = 20)
plt.title("Idades dos não-treineiros")
```

#Desafio05: Comparar as distribuições das provas em inglês espanhol

```
ingles = dados.query('TP_LINGUA==0')
espanhol = dados.query('TP_LINGUA==1')
provas = ["NU_NOTA_CN","NU_NOTA_CH","NU_NOTA_MT","NU_NOTA_LC","NU_NOTA_REDACAO"]
ingles[provas].boxplot(grid=True, figsize= (10,8))
plt.title("Notas das provas - Inglês")
```

```
espanhol[provas].boxplot(grid=True, figsize= (10,8))
plt.title("Notas das provas - Espanhol")
```
